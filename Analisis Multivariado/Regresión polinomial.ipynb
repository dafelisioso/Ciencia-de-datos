Con la función de cess le doy, $n=25$,$x$ que va de 0 a 25  y nuestros valores obtenidos de manera empírica y esta nos regresa el vector de las constantes, con estos valores depues genero mi ploinomio de grado 3 y lo gráfico.

Para sacar el SEE primero sacó el mean squared error y lo multiplico por $n=25$ y ya nos da el error, entre los valores obtenidos y los desados.

Notamos que el error nunca se hace cero,esto se debe por que nos aproxima una función polinomial pero esto no es perfecto y genera los mismos datos, sino que hace una función parecida a los datos obtenidos de manera empirica, pero justo esto se hace por que es muy fácil de manejar los polinomios, por lo que hace que una aproximación a nuestros datos sea mucho más fácil de usar.

```python
def cess(n,x, y):
  V=np.zeros((n,4))
  for i in range(25):
    for j in range(4):
      V[i,j]=i**j
  y2=np.matmul(V.T, y)
  M=np.matmul(V.T,V)
  array_c=la.solve(M,y2)
  return(array_c)
```
