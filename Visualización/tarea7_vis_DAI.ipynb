```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('mpg_cars1.csv')
df.head()
```

```python
df=df[['cty','class']]
df
```

```python

plt.rcParams["figure.figsize"]=(13,10)
plt.rcParams['figure.dpi'] = 80
sns.set(style="whitegrid")
sns.histplot(data = df, x = "cty", kde =True, hue = "class",  alpha=0.7)
sns.kdeplot(data = df, x = "cty",  hue = "class",  alpha=1,linewidth = 3)
plt.title('Autonomía en la ciudad', fontsize=22)
plt.show()
```
