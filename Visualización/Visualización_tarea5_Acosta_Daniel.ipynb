Genere una copia del conjunto de datos y rellene los datos faltantes (NA's) con
un valor de 0 en el caso de datos numéricos usados como identificador la palabra
“desconocido" en el caso de datos tipo cadena de caracteres y en el caso de variables
numéricas use el promedio de los valores de esa columna (p.ej., la edad y la tarifa).

```python
df_copy=df.copy()
df_copy
```

```python
df_copy['body'] = df_copy['body'].fillna(0)
df_copy['cabin'] = df_copy['cabin'].fillna(0)
df_copy['boat'] = df_copy['boat'].fillna(0)
df_copy['home.dest'] = df_copy['home.dest'].fillna(0)
df_copy.head()
```

```python
df_copy['age'] = df_copy['age'].fillna(np.mean(df_copy['age']))
df_copy['fare'] = df_copy['fare'].fillna(np.mean(df_copy['fare']))
df_copy.tail()
```

Finalmente, de los campos “age” y “fare” agregue columnas al conjunto de datos qué
contengan los valores normalizados. Elija la normalización tipo $\frac{x_i-\bar{x_{min}}}{\sigma}$ para el caso de que la
variable tenga una distribución normal y utilice la normalización tipo $\frac{x_i-x_{min}}{x_{max}-x_{min}}$ en cualquier otro caso.


```python
plt.hist(df_copy['age'])

```

```python
plt.hist(df_copy['fare'])
```

Por lo que de sus gráficas podemos decir que la edad sigue más o menos na distribución normal, en cambio la tarifa no

```python
df_copy['age_norm']= (df_copy['age']-df_copy['age'].mean())/df_copy['age'].std()
df_copy['fare_norm']= (df_copy['fare']-df_copy['fare'].min())/(df_copy['age'].max()-df_copy['age'].min())
df_copy.head()
```

```python
Vemos la matrix de distancia hecha por nosotros y la ya implementada
```

```python
from scipy.spatial import distance_matrix
from scipy.cluster import hierarchy
```

```python
pd.DataFrame(distance_matrix(dfm2.values, dfm2.values,p=3), index=dfm2.index, columns=dfm2.index)
```

```python
def dm(a,b,p):

  y=np.zeros((len(a),len(b)))
  for i in range(len(a)):
    for j in range(len(b)):
      y[j][i] =  (sum(np.abs((a[i]-b[j]))**p))**(1/p)
  return(y)
```

```python
dm(dfm2.values, dfm2.values,p=1)
```

Dendograma  normlaizado y nonormalizado

```python
Z=hierarchy.linkage(dfm2,metric='euclidean')
plt.figure()
dn1 = hierarchy.dendrogram(Z)
```
