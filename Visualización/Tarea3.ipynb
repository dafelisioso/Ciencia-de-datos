# Actividad 1

Utilice el módulo scipy.fft de Python para la transformada rápida de Fourier de la siguiente función:

$$f(t)=3sin(7.2 \pi t + 2)+1$$

Grafique la función y visualícela de manera que se pueda ver bien toda la función y sus cambios. Puede usar t=np.linspace(0,T,N) donde T=10 y N=256

```python
import scipy.fft
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft
```

```python
N = 256
T = 10
t = np.linspace(0.0, T, N)
y = 3*np.sin(7.2*np.pi*t+2)+1
yf = fft(y)
plt.plot(yf, color='pink')
plt.plot(y,color='magenta')
```

# Actividad 2

Dado el siguiente código, obtenga la Transformada inverda directa e inversa en 2D así como sus gráficas y visualícelas de tal forma que se puedan ver bien todas las funciones y sus cambios.

```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi

[x,y]= np.meshgrid(np.linspace(0,10,64), np.linspace(0,10,64))
f = np.sin(2*pi*(0.2*x+0.7*y))+np.random.uniform(0,1,x.shape)
plt.pcolor(x,y,f)
```

Graficamos lo anterior en 3D

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, f, cmap="viridis")
plt.show()
```

Aplicamos la transformada de Fourier y lo centramos 

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

xf = np.linspace(0,10,64)
yf = np.linspace(0,10,64)
XF, YF = np.meshgrid(xf, yf)
Z=np.sin(2*pi*(0.2*x+0.7*y))+np.random.uniform(0,1,x.shape)
ZF = scipy.fft.fft2(Z)
ZF = np.fft.fftshift(ZF)

ax.plot_surface(XF, YF, ZF.real, cmap="viridis")
plt.show()
```

Aplicamos la inversa a la transformada de Fourier para obtener la gráfica original

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

xf = np.linspace(0,10,64)
yf = np.linspace(0,10,64)
XF, YF = np.meshgrid(xf, yf)
ZFI = np.fft.fftshift(ZF)
ZFI = scipy.fft.ifft2(ZFI)

ax.plot_surface(XF, YF, ZFI.real, cmap="viridis")
plt.show()
```
