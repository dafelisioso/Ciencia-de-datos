# Ejercicio 1.2

## Elabore una grafica de dispersión que indique el peso de los pacientes (wt_kg) en relación con la edad de los pacientes (age) y el género (gender). Esto es, toda la información solicitada debe observarse en una sola gráfica.

```python
#Detectamos m=1 y f=0 y lo guradamos en el vector genero
genero = []
color = []
for i in range(len(data)):
    g = data["gender"][i]
    if g == "m":
        genero.append(1)
        color.append("pink")
    else:
        genero.append(0)
        color.append("blue")
```

```python
#Graficamos en 3D para que se vea más gráfico
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(xs = data["wt_kg"], ys = data["age"], zs=data["genero"], c = color, alpha = 0.6)
ax.set_xlabel("Peso en kg")
ax.set_ylabel("Edad")
ax.set_zlabel("Sexo")
```

```python
military['year'] = military['date_hospitalisation'].dt.year
military['month'] = military['date_hospitalisation'].dt.month
military['day'] = military['date_hospitalisation'].dt.day
military['dt']=0
military['week']=0
for i in range(len(military)):
    y=military['year'][i]
    m=military['month'][i]
    d=military['day'][i]
    military['dt'][i]=datetime.date(y,m,d)
    military['dt'][i] = datetime.date(military['year'][i],military['month'][i], military['day'][i])
    military['week'][i]=military['dt'][i].isocalendar()[1]
military['year'] = military['year'].astype(int)
```

```python
#Segmentamos por outcome, en este caso como quitamos na, sólo tenemos dos opciones
muertos=data2[data2['outcome']=='Death']
recuperado=data2[data2['outcome']=='Recover']
```

# Para recuperados

```python
#Ordenamos y renombramos indices del renglon
recuperado=recuperado.sort_values(by='date_hospitalisation', ascending=True)
recuperado.index=range(len(recuperado))
recuperado
```

```python
#Transformamos la fecha y creamos nuevas columnas que nos ayudarán a aagrupar
recuperado['year'] = recuperado['date_hospitalisation'].dt.year
recuperado['month'] = recuperado['date_hospitalisation'].dt.month
recuperado['day'] = recuperado['date_hospitalisation'].dt.day
recuperado['dt']=0
recuperado['week']=0
for i in range(len(recuperado)):
    y=recuperado['year'][i]
    m=recuperado['month'][i]
    d=recuperado['day'][i]
    recuperado['dt'][i]=datetime.date(y,m,d)
    recuperado['dt'][i] = datetime.date(recuperado['year'][i],recuperado['month'][i], recuperado['day'][i])
    recuperado['week'][i]=recuperado['dt'][i].isocalendar()[1]
recuperado['year'] = recuperado['year'].astype(int)
recuperado
```

```python
#Separamos recuperados por año y hacemos el groupby por semana
recuperados2014=recuperado[recuperado['year']==2014].groupby('week')
recuperados2014=recuperados2014.size().reset_index(name='counts')


recuperados2015=recuperado[recuperado['year']==2015].groupby('week')
recuperados2015=recuperados2015.size().reset_index(name='counts')

print(recuperados2015)
```
